```python
# pl_df = pd.read_csv('2018-2019-RD.csv')
pl_df = pd.read_csv('https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/2018-2019-RD.csv')
pl_df.head()
print(pl_df.shape)
pl_df.insert(loc=7, column = 'full_time_winner', value = '?')
pl_df.head()
print(pl_df.shape)

lfc_df = pd.read_csv('https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/lfc-2018-2019-player-stats.csv')
lfc_df.insert(loc=0,column='club', value='LFC')
print(lfc_df.shape)
lfc_df.head(10)

mcfc_df = pd.read_csv('https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/mcfc-2018-2019-player-stats.csv')
mcfc_df.insert(loc=0,column='club', value='MCFC')
print(mcfc_df.shape)
mcfc_df.head(10)
```

```python
teams = ['Liverpool','Man City']

# Away match DFs
home = pl_df[pl_df.home_team.isin(teams)]
print(home.shape)
mc_home = pl_df[pl_df.home_team == 'Man City']
print(mc_home.shape)
lfc_home = pl_df[pl_df.home_team == 'Liverpool']
print(lfc_home.shape)

# Home match DFs
away = pl_df[pl_df.away_team.isin(teams)]
print(away.shape)
mc_away = pl_df[pl_df.away_team == 'Man City']
print(mc_away.shape)
lfc_away = pl_df[pl_df.away_team == 'Liverpool']
print(lfc_away.shape)

# Combined LFC and MCFC match DF
liv_mc = pd.concat([home,away], ignore_index=True)
print(liv_mc.shape)
liv_mc.head(10)

# City's entire match DF
mcfc_matches = pd.concat([mc_home, mc_away], ignore_index=True)
# mc_matches.to_csv('mcfc_matches.csv')
print(mc_matches.shape)

# Liverpool's entire match DF
lfc_matches = pd.concat([lfc_home, lfc_away], ignore_index=True)
# lfc_matches.to_csv('lfc_matches.csv')
print(lfc_matches.shape)
```

```python
for t in liv_mc.itertuples():
    idx = t[0]
    winner = t.full_time_result
    h = t.home_team
    a = t.away_team
    
    if winner == 'A':
        liv_mc['full_time_winner'][idx] = str(a)
    
    elif winner == 'H':
        liv_mc['full_time_winner'][idx] = str(h)
        
liv_mc.head(10)
```

```python
lfc_goals_t = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/lfc_matches.csv"},
  "mark":{"type":"trail", "color":"red", "tooltip":True},
  "encoding":{
    "x":{"field":"date", "type":"ordinal", "title":"Match Date"},
    "y":{"field":"lfc_goals", "type":"quantitative", "title":"Liverpool Goals Scored, by Match"}
  }
})

city_goals_t = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/mcfc_matches.csv"},
  "mark":{"type":"trail", "color":"skyblue", "tooltip":True},
  "encoding":{
    "x":{"field":"date", "type":"ordinal", "title":"Match Date"},
    "y":{"field":"mcfc_goals", "type":"quantitative", "title":"Man City Goals Scored, by Match"}
  }
})

lfc_goals_t | city_goals_t
```

```python
# TEAM GOALS

brush = alt.selection_single(encodings=["x","y"])

lfc_goals = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/lfc_matches.csv"},
  "mark":{"type":"bar", "color":"red", "tooltip":True},
  "encoding":{
    "x":{"field":"date", "type":"ordinal", "title":"Match Date"},
    "y":{"field":"lfc_goals", "type":"quantitative", "title":"Liverpool Goals Scored, by Match"}
  }
}).add_selection(
    brush
)

city_goals = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/mcfc_matches.csv"},
  "mark":{"type":"bar", "color":"skyblue", "tooltip":True},
  "encoding":{
    "x":{"field":"date", "type":"ordinal", "title":"Match Date"},
    "y":{"field":"mcfc_goals", "type":"quantitative", "title":"Man City Goals Scored, by Match"}
  }
}).add_selection(
    brush
)

lfc_goals | city_goals
```

```python
# REFS

lfc_refs = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/lfc_matches.csv"},
  "mark":{"type":"point", "color":"red", "tooltip":True},
  "encoding":{
    "x":{"field":"date", "type":"ordinal", "title":"Match Date"},
    "y":{"field":"referee", "type":"nominal", "title":"Referee"}
  }
}).transform_filter(
    brush
)

city_refs = alt.Chart.from_dict({
  "data":{"url":"https://raw.githubusercontent.com/rdubnic2/rdubnic2.github.io/main/mcfc_matches.csv"},
  "mark":{"type":"point", "color":"navy", "tooltip":True},
  "encoding":{
    "x":{"field":"date", "type":"ordinal", "title":"Match Date"},
    "y":{"field":"referee", "type":"nominal", "title":"Referee"}
  }
}).transform_filter(
    brush
)
```
